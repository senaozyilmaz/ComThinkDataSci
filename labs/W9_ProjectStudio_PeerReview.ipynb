## Peer Review Checklist (copy/paste into your notebook submission)

### A) Reproducibility (must-pass)
- [ ] Notebook runs from top without manual fixes
- [ ] Data loading is reproducible (URL or file included)
- [ ] All variables used are defined (no hidden state)
- [ ] Outputs/figures have titles and axis labels
- [ ] Randomness is controlled (seed) where used

### B) Data & curation
- [ ] Clear description of dataset and target definition (how `disease` is defined)
- [ ] Missingness handled with an explicit policy (drop/impute/flag) and justification
- [ ] Any recoding is documented (e.g., `cp` labels)
- [ ] At least one sanity check / quality gate exists (assertions)

### C) EDA quality
- [ ] At least 6 plots, with a clear purpose (not decorative)
- [ ] EDA includes distributions + relationships + subgroup or stratified view
- [ ] Outliers discussed (error vs rare-but-real)

### D) Inference / testing
- [ ] Hypotheses written as H0/H1
- [ ] Test choice matches variable types
- [ ] Report includes effect size + uncertainty (CI) (not only p-value)
- [ ] Assumptions stated and checked (at least 2)

### E) Critical thinking
- [ ] Identifies at least one confounder and attempts stratification or control
- [ ] Avoids causal claims (unless study design supports it)
- [ ] Conclusion matches evidence strength (no overclaiming)

### F) Communication
- [ ] One clear final claim in plain language
- [ ] One paragraph limitations + what would change your mind
