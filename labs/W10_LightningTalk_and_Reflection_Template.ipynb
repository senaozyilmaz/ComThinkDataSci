## Talk outline (fill these sections)
) Title + your main claim (1 sentence)

Title: Maximum Heart Rate and Exercise-Induced Angina Are Associated with Heart Disease Risk

Main claim (1 sentence, non-causal unless justified): After adjusting for age and sex, lower maximum heart rate and presence of exercise-induced angina are each independently associated with higher odds of heart disease, but these associations are not proven to be causal.

2) Why do we care (bioengineering relevance, 2–3 sentences)

Clinicians need simple, non-invasive indicators to screen for coronary artery disease. Understanding how exercise test variables (heart rate, angina, ST depression) relate to disease risk can help design better diagnostic algorithms and wearable monitoring systems. Bioengineers use such associations to build predictive models for remote cardiac risk assessment.

3) Data (what, where from, sample size, target definition)

Source: UCI Machine Learning Repository – Cleveland Heart Disease dataset

Rows (n): 303 original, 267 after dropping missing values on used features

Target definition (disease): Presence of heart disease (num > 0), where num is 0 (no disease) vs 1,2,3,4 (vessel narrowing percentages)

Key caveats: Observational data from a single hospital in the 1980s; missing values in ca and thal columns; no information on medications, lifestyle, or genetic risk.

4) Methods (what you did)

Cleaning policy: Replaced "?" with NaN, converted numeric columns, dropped rows with missing values in analysis variables (listwise deletion), created binary disease outcome (num > 0).

EDA highlights: (added after peer review) Distribution plots showed that diseased patients are older, have lower thalach, higher oldpeak, and more frequent exang. Correlation matrix indicated thalach and oldpeak are moderately correlated with disease.

Test/model used: Logistic regression (baseline: age + sex; extended: age + sex + thalach + oldpeak + exang). Reported odds ratios with 95% confidence intervals.

5) Results (one plot + one statistic)

Figure: Side-by-side boxplot of maximum heart rate (thalach) by disease status (0 = no disease, 1 = disease), with group sizes printed: n_no_disease = 138, n_disease = 129. Title: "Maximum Heart Rate is Lower in Patients with Heart Disease". X-axis: Disease (0 / 1). Y-axis: Thalach (beats per minute).

Effect size: Odds ratio for thalach = 0.966 (each additional bpm reduces odds of disease by about 3.4%)

CI: 95% confidence interval for thalach OR = [0.947, 0.985]

p-value (if used): p < 0.001 (Wald test)

6) Limitations + what would change your mind

Limitations: Unmeasured confounding (physical fitness, smoking, family history) could explain the observed associations. Listwise deletion of missing cases may bias results if missingness is not random. The dataset is small and from a single clinical site, limiting generalizability.

What would change my mind: A randomized controlled trial showing that increasing maximum heart rate (via exercise training) does not reduce heart disease risk, or a large multi-center observational study that finds no association after adjusting for fitness and lifestyle.

7) Next step (what data or experiment would you do next?)

Collect longitudinal data with baseline measurements of cardiorespiratory fitness (VO2 max), physical activity, and smoking status. Also include a validation cohort from a different hospital or country. If possible, run a small pilot intervention study where sedentary patients undergo supervised exercise and we measure changes in thalach and subsequent cardiac events.

## Reflection prompt (submit 5–8 sentences)
1. One assumption you made that could be wrong:
   I assumed that linearity in the logit holds for continuous predictors like thalach and oldpeak. If the relationship is U-shaped or threshold-based, the odds ratio interpretation would be misleading. I did not formally test this with splines or Box-Tidwell.
2. One thing you did to reduce bias or overclaiming:
   I explicitly avoided causal language in my main claim and interpretations, using phrases like "associated with" instead of "causes". I also identified physical fitness as an unmeasured confounder and noted that listwise deletion may introduce bias if missing data are not random.
3. One thing you would improve if you had another week:
   I would add multiple imputation for missing values instead of dropping rows, check all logistic regression assumptions (multicollinearity via VIF, linearity via splines), and create a stratified analysis by sex to see if associations differ between men and women. I would also seek an external validation dataset to test whether the odds ratios generalize.
